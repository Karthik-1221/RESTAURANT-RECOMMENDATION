```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
df['Switch to order menu'] = labelencoder.fit_transform(df['Switch to order menu'])
df['Has Table booking'] = labelencoder.fit_transform(df['Has Table booking'])
df['Has Online delivery'] = labelencoder.fit_transform(df['Has Online delivery'])
df['Is delivering now'] = labelencoder.fit_transform(df['Is delivering now'])
df.head()
```

```python
#ORDINAL ENCODING FOR RATING COLOR COLUMN

from sklearn.preprocessing import OrdinalEncoder
ordinalencoder = OrdinalEncoder(categories=[['Orange', 'White', 'Yellow','Green','Dark Green','Red']])
df['Rating color'] = ordinalencoder.fit_transform(df[['Rating color']])
df.head()
```

```python
#ORDINAL ENCODING FOR RATING TEXT COLUMN
ordinalencoder = OrdinalEncoder(categories=[['Not rated','Poor','Average','Good','Very Good','Excellent']])
df['Rating text'] = ordinalencoder.fit_transform(df[['Rating text']])
df.head()
```

```python
#I AM USING MINMAX SCALAR TO SCALE THE VALUES BETWEEN 0 AND 1 (MINMAX NORMALIZATION)

from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
df[['Country Code','Average Cost for two','Price range','Restaurant ID','Longitude','Latitude']] = minmax_scaler.fit_transform(df[['Country Code','Average Cost for two','Price range','Restaurant ID','Longitude','Latitude']])
df.head()
```

```python
df_dummies = pd.get_dummies(df, columns=['Cuisines'], drop_first=True)
df_dummies.head()
```

```python
df_numeric = df_dummies *1
df_numeric.head()
```

```python
df_dummies.info()
```

```python
#CONTENT BASED FILTERING
```

```python
#defining criteria  and using cosine similarity  for similarity calculation  
```

```python
from sklearn.metrics.pairwise import cosine_similarity

criteria = ["Cuisines_Vietnamese", 'Price range']

criteria_df = df_numeric[criteria]
similarity_matrix = cosine_similarity(criteria_df)
```

```python
criteria_df
```

```python
similarity_matrix
```

```python
def get_recommendations(restaurant_index, num_recommendations=5):
    sim_scores = list(enumerate(similarity_matrix[restaurant_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_indices = [i[0] for i in sim_scores[1:num_recommendations+1]]

    return df_numeric.iloc[sim_indices]

# Example 
sample_restaurant_index = 1
recommendations = get_recommendations(sample_restaurant_index)
recommendations
```

```python
 #Test the Recommendation System
```

```python
sample_criteria = {'Cuisines_Vietnamese': 1, 'Price range': 2}
user_df = pd.DataFrame([sample_criteria])
user_similarity = cosine_similarity(user_df, criteria_df)
user_sim_indices = list(enumerate(user_similarity[0]))
user_sim_indices = sorted(user_sim_indices, key=lambda x: x[1], reverse=True)
recommended_restaurants = df_numeric.iloc[[i[0] for i in user_sim_indices[:5]]]

recommended_restaurants
```
